# Implementing a Basic Model to try and Diagnose MS from MRI scans

## Get the dataset from Kaggle

```python
import os
import kagglehub

# Set custom download directory
os.environ["KAGGLEHUB_CACHE"] = "../data/raw"

# Now download
if not os.listdir("../data/raw"):
    path = kagglehub.dataset_download("buraktaci/multiple-sclerosis")
    print("Dataset downloaded to:", path)
else:
    print("Dataset already exists in ../data/raw")
```

## Assigning Path Vars

```python
# Assigning mri image dataset paths to variables
datasets = {}
datasets["control_axial"] = ["../data/raw/datasets/buraktaci/multiple-sclerosis/versions/1/MS/Control Axial_crop"]
datasets["control_sagittal"] = ["../data/raw/datasets/buraktaci/multiple-sclerosis/versions/1/MS/Control Saggital_crop"]
datasets["ms_axial"] = ["../data/raw/datasets/buraktaci/multiple-sclerosis/versions/1/MS/MS Axial_crop"]
datasets["ms_sagittal"] = ["../data/raw/datasets/buraktaci/multiple-sclerosis/versions/1/MS/MS Saggital_crop"]

# Checking number of files in each dataset as well as total number of files
total = 0
for key,value in datasets.items():
    current = len(os.listdir(value))
    print(f"Number of files in {key}: {current}")
    total += current

print(f"Total number of files: {total}")
```
